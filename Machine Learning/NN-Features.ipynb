```python
# Librerias necesarias para diseñar la red neuronal y analizar los datos
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,  r2_score
from scipy.stats import describe
%pylab inline
```

```python
# Cargar el conjunto de datos
dataset = pd.read_csv('data_features.csv')
print(dataset.shape)
dataset.tail()
```

```python
# Remover la columna que contiene la informacion de la temperatura crítica
target_feature = ["critical_temp"]
features = list(set(list(dataset.columns))-set(target_feature))

#Normalizar los valores de los datos
dataset[features] = dataset[features]/dataset[features].max()

x = dataset[features].values
y = dataset[target_feature].values

plt.figure(figsize = (8,6))
plt.hist(y, bins = 30, edgecolor = 'black', alpha = 0.5)
plt.xlim(xmin = -5, xmax = 150)
plt.xlabel('Temperatura crítica (K)', fontsize = 14)
plt.ylabel('Número de compuestos', fontsize = 14)
plt.savefig('Temperaturas_hist.png')
```

```python
describe(y)
```

```python
# Seleccionar los datos que se utilizan para entrenar la red (80%) y para evaluar (20%)
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
print(xtrain.shape, ytrain.shape)
print(xtest.shape, ytest.shape)

# Diseño de la red neuronal
def Model():
    
    # Esta red neuronal tiene 10 capas, de las cuales 8 son ocultas, la funcion de activacion en todas las capas es RELU
    model = keras.Sequential([
    layers.Dense(264, input_dim = 81, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(4, activation='relu'),
    layers.Dense(1, activation='relu')
    ])
    
    #Para optimizar el algoritmo de propagación se emplea el método de Adam de Keras
    optimizer = keras.optimizers.Adam(learning_rate = 0.001)
    
    # La función de costo que se utiliza es el error cuadrático medio
    model.compile(loss = 'mean_squared_error', optimizer = optimizer, metrics = ['mean_squared_error'])
    return model

model = Model()
model.summary()
# Se entrena la red durante 1000 epocas con un tamaño de batch de 64
history = model.fit(xtrain, ytrain, epochs = 1000 , batch_size = 64, verbose = 1)
```

```python
# Se realizan las predicciones con los datos de evaluación y se calcula el valor de R^2 para saber la confiabilidad
prediction = model.predict(xtest)
print(r2_score(ytest, prediction))

plt.figure(figsize = (5,5))
plt.plot(ytest, ytest, linewidth = 2, color = 'k')
plt.scatter(ytest, prediction, edgecolors='k', color = 'tomato', alpha = 0.6)
plt.ylabel('T predicción (K)', fontsize = 14)
plt.xlabel('T de prueba (K)', fontsize = 14)
#plt.savefig('regresion-features.png')
```
